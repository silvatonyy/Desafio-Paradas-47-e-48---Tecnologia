```python
import pandas as pd

df = pd.read_csv('servidores_da_rede_federal.csv', sep=';', encoding='latin-1')

coluna = 'Instituição'
dado = 'IF BAIANO'

contagem = df[coluna].value_counts().get(dado, 0)

print(f'O dado "{dado}" aparece {contagem} vezes na coluna "{coluna}",')
```

Criando Gráficos com Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('servidores_da_rede_federal.csv', sep=';', encoding='latin-1')

plt.bar(df['Instituição'],df['Quantidade'])

plt.title('Servidores da rede por Estado')
plt.xlabel('IF-BAIANO')
plt.ylabel('Graduação')

plt.show()


```
